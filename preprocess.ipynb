```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Absenteeism_data.csv')
data.head()
```

```python
data = data.copy()
data=data.drop(['ID'], axis=1)
```

```python
data.head()
```

# Reason for Absence column

split the column into 4 groups as dummy variables 

```python
data['Group 1']=data['Reason for Absence'].map(lambda x:1 if(x>0 and x<=14) else 0)
data['Group 2']=data['Reason for Absence'].map(lambda x:1 if(x>14 and x<=17) else 0)
data['Group 3']=data['Reason for Absence'].map(lambda x:1 if(x>17 and x<=21) else 0)
data['Group 4']=data['Reason for Absence'].map(lambda x:1 if(x>21 and x<=28) else 0)

```

```python
## removing the 'Reason for Absence' column
data=data.drop(['Reason for Absence'],axis=1)
data.head()
```

# checkpoint

```python
df_reason_mod =  data.copy()
```

# date column
extract the month value  and the day of week value from the 'date' column.
then, delete 'date' column.

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod['Month value'] = df_reason_mod['Date'].map(lambda date: date.month)
```

```python
df_reason_mod['Day value'] = df_reason_mod['Date'].map(lambda date: (date.weekday()+1))
```

```python
df_reason_mod=df_reason_mod.drop(['Date'], axis=1)
df_reason_mod.head()
```

# checkpoint

```python
df_reason_date_mod = df_reason_mod.copy()
```

# education column
turn the data from 'Education column' into a binary column, by mapping 0 to the value of 1, and the value of 1 to the rest of the values in the columns.

```python
df_reason_date_mod['Education'] =df_reason_date_mod['Education'].map(lambda x : 0 if(x<=1) else 1)
```

```python
data_preprocessed = df_reason_date_mod.copy()
```

```python
data_preprocessed.head()
data_preprocessed.to_csv('df_preprocesed.csv',index=False)
```
